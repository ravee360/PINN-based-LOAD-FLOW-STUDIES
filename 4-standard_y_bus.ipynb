```python
import pandapower as pp
import pandapower.networks as pn
import numpy as np
import pandas as pd

# Load IEEE-33 bus system
net = pn.case33bw()

# Run AC power flow (needed to build internal structures)
pp.runpp(net)

# Extract Y-bus (complex sparse matrix)
Ybus_sparse = net._ppc["internal"]["Ybus"]

# Convert to dense NumPy array
Ybus = Ybus_sparse.toarray()

print("Y-bus shape:", Ybus.shape)
```

```python
import pandapower as pp
import pandapower.networks as pn
import pandas as pd

# Load IEEE-33 bus system
net = pn.case33bw()

# Run power flow
pp.runpp(net, numba=False)

# Extract Y-bus
Ybus = net._ppc["internal"]["Ybus"].toarray()

# Separate real and imaginary parts
G = Ybus.real   # Conductance matrix
B = Ybus.imag   # Susceptance matrix

# Save to CSV files (same folder as script)
pd.DataFrame(G).to_csv("Ybus_G_real_ieee33.csv", index=False)
pd.DataFrame(B).to_csv("Ybus_B_imag_ieee33.csv", index=False)

print("Y-bus saved successfully!")
```
