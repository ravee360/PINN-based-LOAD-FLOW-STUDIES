```python
df.drop(columns=[df.columns[0]], inplace=True)
```

```python
import re

order = ["p", "q", "pl", "ql", "v"]
df = df[sorted(
    df.columns,
    key=lambda x: (int(re.search(r'\d+', x).group()), order.index(re.search(r'^[a-z]+', x).group()))
)]
```

```python
df.head()
```

```python
list(df.columns)
```

```python

# step 1
cols = df.columns.tolist()

# step 2: extract numeric suffixes
nums = sorted({re.findall(r'\d+', c)[0] for c in cols})

# step 3: expected features
req = ["p", "q", "pl", "ql", "v"]

# step 4: find valid bus numbers
valid_nums = [n for n in nums if all(f"{k}{n}" in cols for k in req)]

# step 5: build final column order
keep_cols = []
for n in valid_nums:
    keep_cols += [f"p{n}", f"q{n}", f"pl{n}", f"ql{n}", f"v{n}"]

# step 6: filter dataframe
df = df[keep_cols]
```

```python
# Suppose df is already loaded
cols = df.columns.tolist()

# Extract prefix and number
import re
parsed = [(re.match(r'([a-z]+)(\d+)', c).groups()[0], int(re.match(r'([a-z]+)(\d+)', c).groups()[1]), c) for c in cols]

# Define desired prefix order
prefix_order = {'p': 0, 'q': 1, 'pl': 2, 'ql': 3, 'v': 4}

# Sort by number first, then by prefix order
sorted_cols = [c for _, _, c in sorted(parsed, key=lambda x: (x[1], prefix_order[x[0]]))]

# Reorder dataframe
df = df[sorted_cols]
```
